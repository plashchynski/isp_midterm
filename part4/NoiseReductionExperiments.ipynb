Low-pass filter

```python
from scipy.signal import butter, filtfilt
```

```python
def lowpass_filter(audio, sr, cutoff=1000, order=4):
    normalized_cutoff = cutoff / (sr / 2)

    b, a = butter(order, normalized_cutoff, btype='low', analog=False, output='ba')
    return filtfilt(b, a, audio)
```

```python
Audio(lowpass_filter(original_audio, sr=sr, cutoff=3000), rate=sr)
```

# spleeter

```python
from spleeter.separator import Separator

# Using embedded configuration.
separator = Separator('spleeter:2stems')
```

```python
# Use audio loader explicitly for loading audio waveform :
from spleeter.audio.adapter import AudioAdapter

audio_loader = AudioAdapter.default()

waveform, sr = audio_loader.load("Audio_Files/EN/your_sentence1.wav")

# Perform the separation :
prediction = separator.separate(waveform)
```

```python
Audio(waveform[:, 0], rate=int(sr))
```

```python
Audio(prediction['vocals'][:, 0], rate=int(sr))
```
